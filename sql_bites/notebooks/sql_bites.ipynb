The HAVING keyword is a bit like WHERE. However, WHERE can't be used with aggregate functions, so we can use HAVING instead. For example, we can refine the last query above a little further:

```python
%%sql
SELECT "City Name", "Variety", COUNT(*) 
FROM us_pumpkin_prices 
GROUP BY "City Name", "Variety" 
HAVING MAX("High Price") > 300 
ORDER BY "City Name" ASC;
```

Write a query that returns the varieties of pumpkins in different cities, grouped by "City Name" that are having a minimum price of 5, and ordered by "Variety".

```python
%%sql
SELECT "City Name", "Variety"
FROM us_pumpkin_prices 
GROUP BY "City Name", "Variety"
HAVING MIN("Low Price") >= 5
ORDER BY "Variety" ASC;
```

Next I will try to get a table with a count of athletes per country.

```python
%%sql
SELECT country, COUNT(*) AS athletes_per_country
FROM athletes
GROUP BY country
ORDER BY country
LIMIT 10
```

I will make a table with count of medals per country.

```python
%%sql
SELECT country, COUNT(*) AS medals_per_country
FROM medals
GROUP BY country
ORDER BY country
LIMIT 10
```

I will Join these tables to show 3 columns.

```python
%%sql 
SELECT table_a.country AS country, athletes_per_country, medals_per_country
FROM (
    SELECT country, COUNT(*) AS athletes_per_country
FROM athletes
GROUP BY country
) AS table_a
JOIN (
    SELECT country, COUNT(*) AS medals_per_country
FROM medals
GROUP BY country
) AS table_b
ON table_a.country = table_b.country
LIMIT 10
```

Now I will add the final column and order.

```python
%%sql 
SELECT table_a.country AS country, athlete_count, medal_count, round(
    CAST((CAST(medal_count AS float)/ athlete_count)AS numeric)
    , 2) AS medals_per_athlete
FROM (
    SELECT country, COUNT(*) AS athlete_count 
FROM athletes
GROUP BY country
) AS table_a
JOIN (
    SELECT country, COUNT(*) AS medal_count 
FROM medals
GROUP BY country
) AS table_b
ON table_a.country = table_b.country
ORDER BY medals_per_athlete DESC
LIMIT 5
```

Paula wanted to write a query that generates a table showing the average age of competitors winning each medal - gold, silver or bronze. Her query is below - does it work as intended? HINT: The answer is no :)

```python
%%sql
SELECT subquery_result.medal, SUM(subquery_result.age) as average_age
FROM (
  SELECT
    medals.medal,
    DATE_PART('year', CURRENT_DATE) - DATE_PART('year', athletes.date_of_birth) as age
    FROM athletes INNER JOIN medals
    ON athletes.name = medals.name
) subquery_result
GROUP BY medal;
```

The average age is much too high so there must be an issue calculating it. I notice looking at the code that the sum of age is being used rather than the average so I will change this first. 

```python
%%sql
SELECT subquery_result.medal, AVG(subquery_result.age) as average_age
FROM (
  SELECT
    medals.medal,
    DATE_PART('year', CURRENT_DATE) - DATE_PART('year', athletes.date_of_birth) as age
    FROM athletes INNER JOIN medals
    ON athletes.name = medals.name
) subquery_result
GROUP BY medal;
```

This appears to give much more sensible answers. However displaying these answers to such high precision is misleading as the calculated age for each athlete is an integer. Below I have corrected the precision to 0 decimal places. 

```python
%%sql
SELECT subquery_result.medal, round(CAST(AVG(subquery_result.age) AS numeric), 0) as average_age
FROM (
  SELECT
    medals.medal,
    DATE_PART('year', CURRENT_DATE) - DATE_PART('year', athletes.date_of_birth) as age
    FROM athletes INNER JOIN medals
    ON athletes.name = medals.name
) subquery_result
GROUP BY medal;
```

This is more honest precision in terms of how the ages were calculated but could be better. 

My age is not the current year minus my birth year, the exact current and birth date matters. I can include months and days to make age a float rather than integer and be more accurate. 

I have displayed 5 decimal places below to compare to the years only values.

```python
%%sql
SELECT subquery_result.medal, round(CAST(AVG(subquery_result.age) AS numeric), 5) as average_age
FROM (
  SELECT
    medals.medal,
    (DATE_PART('year', CURRENT_DATE) - DATE_PART('year', athletes.date_of_birth) 
     + ((DATE_PART('month', CURRENT_DATE) - DATE_PART('month', athletes.date_of_birth))/12)
     + ((DATE_PART('day', CURRENT_DATE) - DATE_PART('day', athletes.date_of_birth))/365.25))
    as age
    FROM athletes INNER JOIN medals
    ON athletes.name = medals.name
) subquery_result
GROUP BY medal;
```

Below I round to their precision to 1 decimal place. 

```python
%%sql
SELECT subquery_result.medal, round(CAST(AVG(subquery_result.age) AS numeric), 1) as average_age
FROM (
  SELECT
    medals.medal,
    (DATE_PART('year', CURRENT_DATE) - DATE_PART('year', athletes.date_of_birth) 
    + ((DATE_PART('month', CURRENT_DATE) - DATE_PART('month', athletes.date_of_birth))/12)
    + ((DATE_PART('day', CURRENT_DATE) - DATE_PART('day', athletes.date_of_birth))/365.25))
    as age
    FROM athletes INNER JOIN medals
    ON athletes.name = medals.name
) subquery_result
GROUP BY medal;
```
